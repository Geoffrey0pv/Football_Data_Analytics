## **3. A member of the coaching staff claims that tall goalkeepers (GK) perform better overall. What can you conclude from a descriptive analysis?**

**(Note: use three height groups as a definition:**

**i) less than 185 cm.**

**ii) between 185 and 192 cm.**

**iii) more than 192 cm.**

* First, we filter the dataset with only goalkeepers

```python
goalkeepers = dataset[(dataset['position'] == 'GK')]
```

```python
goalkeepers
```

* Then we create three categories to difference goalkeepers because: 

    * **Comparability**	It lets you group goalkeepers into meaningful physical categories.
    * **Variance Analysis**	ANOVA needs distinct groups. We can't run ANOVA on continuous height directly.
    * **Interpretability**	we can explain: "Taller goalkeepers (above 192 cm) vs Medium (185–192 cm) vs Short (below 185 cm)."
    * **Real-world application** Scouts and coaches think in categories: short, average, tall. This mirrors reality

```python
goalkeepers['height_category'] = pd.cut(goalkeepers['height_cm'],
                                bins=[0, 185, 192, goalkeepers['height_cm'].max()],
                                labels=['Bajo', 'Medio', 'Alto'])
```

```python
goalkeepers_filter = goalkeepers[['height_category', 'player_name', 'position', 'overall', 'height_cm']]
```

```python
goalkeepers_filter
```

* **Diagram showing the relationship between height and average**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=goalkeepers_filter, x='height_cm', y='overall', hue='position') 
plt.title('Relación entre Estatura y Overall')
plt.xlabel('Estatura (cm)')
plt.ylabel('Overall')
plt.grid(True)
plt.show()
```

* **Descriptive Statistics for height category by overall points**

```python
goalkeepers.groupby('height_category')['overall'].describe()
```

* **Boxplots**

We are going to plot box diagrams to see the average for each category and how many outliers they can have 

```python
sns.boxplot(data=goalkeepers_filter, x='height_category', y='overall')
plt.title('Goalkeeper Overall Rating by Height Category')
plt.show()
```

Now we can see such as the categories has a lot of outliers above average, especially the categories "medio" and "alto" it seams have most outliers, and "bajo" category it seams have the lowest average

* **Histograms**

We are going to plot histograms to see how the data are distributed 

```python
for category in ['Bajo', 'Medio', 'Alto']:
    plt.figure(figsize=(10, 8))
    sns.histplot(goalkeepers_filter[goalkeepers_filter['height_category'] == category]['overall'], kde=True)
    plt.title(f'Overall Distribution - {category} Goalkeepers')
    plt.xlabel('Overall')
    plt.show()
```

The distributions are close to normal (especially for "Alto" and "Medio"), so ANOVA is appropriate.

* **Statistical Test: ANOVA**

We are going to use a ANOVA (Analysis of Variance) because:

It allows us to statistically compare the means of three or more independent groups (in our case: Short, Medium, Tall goalkeepers) to determine if at least one group has a significantly different mean overall rating.

```python
short = goalkeepers_filter[goalkeepers_filter['height_category'] == 'Bajo']['overall']
medium = goalkeepers_filter[goalkeepers_filter['height_category'] == 'Medio']['overall']
tall = goalkeepers_filter[goalkeepers_filter['height_category'] == 'Alto']['overall']

f_stat, p_value = f_oneway(short, medium, tall)
print("F-statistic:", f_stat)
print("p-value:", p_value)
if p_value < 0.05:
    print("At least one group has a significantly different overall.")
else:
    print(" No evidence of a significant difference.")
```

* **Conclusion**

An ANOVA test was conducted to assess whether goalkeepers of different height categories (Bajo, Medio, Alto) differ significantly in their overall performance ratings.
The results revealed a statistically significant difference between at least one of the groups (F = 26.847353666658954, p < 0.000001).
On average, taller goalkeepers have higher overall ratings (64.75) compared to medium (64.48) and especially short goalkeepers (62.09).
This supports the idea that height is positively associated with goalkeeper performance, at least as measured in this dataset.

## **4. An analyst claims that **left-footed players** have **better performance (overall)** than right-footed ones. What can you say about this statement based on descriptive analysis**?



First, we need to know how the data is distributed and meet the statics measuares such as mean overall club these players have

```python
lefties = dataset[dataset['pref_foot'] == 'Left']
righties = dataset[dataset['pref_foot'] == 'Right']
```

* **Descriptive Statistics for prefeer foot by overall points**

```python
print("Left-footed players:")
print(lefties['overall'].describe())

print("\nRight-footed players:")
print(righties['overall'].describe())
```

Here we can see that the data of the preferred foot is not balancing, there are most football players prefer their right foot and Left-footed players have a slightly higher average overall rating (~0.72 points higher), and a slightly narrower distribution (less spread in scores).

* **Bloxplots**

```python
sns.boxplot(data=dataset, x='pref_foot', y='overall')
plt.title("Overall Performance by Preferred Foot")
plt.show()
```

* the boxplot confirms:

    * Left-footed players have a slightly higher median and upper quartile.
    * Outliers exist in both groups.
    * Distributions overlap heavily

* **Histograms**

We are going to plot histograms to see how the data are distributed 

```python
sns.histplot(lefties['overall'], kde=True, color='blue', label='Left', stat='density')
sns.histplot(righties['overall'], kde=True, color='green', label='Right', stat='density')
plt.legend()
plt.title("Overall Distribution: Left vs Right-footed Players")
plt.show()
```

The histogram shows both distributions are nearly normal and very similar in shape.

* **Statistical Test: T-test**

* In this case we use the independent two-sample t-test because:

    * We wanted to compare the means of two groups (left-footed vs. right-footed players).

    * We overall scores are approximately normally distributed (as seen in your histograms).

    * We had two independent samples: no player appears in both groups.

    * We wanted to know if the observed difference in means is statistically significant (not just due to randomness).

```python
t_stat, p_value = ttest_ind(lefties['overall'], righties['overall'], equal_var=False)
print("T-statistic:", t_stat)
print("P-value:", p_value)
if p_value < 0.05:
    print("There is a significant difference between left-footed and right-footed players.")
else:
    print("No significant difference between left-footed and right-footed players.")
```

Even if the difference in means is small (~0.72), the huge sample size (22,000+ players) 
gives this result high statistical power — that's why the test detects even small effects.

* **Conclusion:** 

A two-sample t-test was performed to determine whether left-footed players have significantly better overall performance ratings than right-footed players.
The results show a statistically significant difference (T = 6.87, p < 0.00001), with left-footed players having a slightly higher average rating (65.77 vs. 65.05).
Although the difference is numerically small (~0.72), it is statistically meaningful given the large sample size.


Therefore, we support the analyst's claim that left-footed players perform better on average, according to the overall rating

## **5. A scout claims that the **best players**, according to their **overall performance rating**, play in clubs that belong to the **“Top 5 leagues”** (England, Spain, Germany, Italy, and France). What can you say about these two claims based on descriptive analysis?**

We define top 5 leagues

```python
top5_leagues = ['England', 'Spain', 'Germany', 'Italy', 'France']
```

Then we divided the group into two categories

```python
top5 = dataset[dataset['club_country'].isin(top5_leagues)]
non_top5 = dataset[~dataset['club_country'].isin(top5_leagues)]
```

* **Descriptive Statistics for overall**

```python
print("Top 5 leagues - Overall:")
print(top5['overall'].describe())

print("\nNon-Top 5 leagues - Overall:")
print(non_top5['overall'].describe())
```

* **Boxplot**

We are going to plot box diagrams to see the average for this two categories and how many outliers they can have 

```python
top5['league_group'] = 'Top 5'
non_top5['league_group'] = 'Other'

combined = pd.concat([top5, non_top5])

sns.boxplot(data=combined, x='league_group', y='overall')
plt.title("Overall Ratings: Top 5 vs Other Leagues")
plt.show()
```

* **Histograms**

We are going to plot histograms to see how the data are distributed 

```python
sns.histplot(top5['overall'], label='Top 5', kde=True, stat='density')
sns.histplot(non_top5['overall'], label='Other', kde=True, stat='density')
plt.legend()
plt.title("Distribution of Overall Ratings")
plt.show()
```

* **Statistical Test: T-test**

```python
t_stat, p_value = ttest_ind(top5['overall'], non_top5['overall'], equal_var=False)
print(f"T-statistic: {t_stat:.3f}, p-value: {p_value:.5f}")

if p_value < 0.05:
    print("There is a significant difference in overall ratings between Top 5 and Other leagues.")
else:
    print("No significant difference in overall ratings between Top 5 and Other leagues.")
```

* **Conclusion**

A descriptive and statistical analysis was conducted to evaluate the scout's claim that the best players play in the Top 5 European leagues (England, Spain, Germany, Italy, France).
Players from these leagues have a higher average overall rating (66.84) compared to those from other leagues (64.28).
A two-sample t-test confirmed that this difference is statistically significant (T = 26.52, p < 0.00001), supporting the scout’s claim.
The data shows that Top 5 leagues tend to attract or develop players with higher performance ratings, according to the overall metric used in this dataset.
